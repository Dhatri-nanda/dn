```python
import numpy as np
from numpy import linalg 

def MGS(A):
  n,n = A.shape
  AT = np.transpose(A)
  Q_T = np.zeros((n,n))
  Q_t_T = np.copy(AT)
  R = np.zeros((n,n))
  for i in range(0, n):
    Q_T[i] = Q_t_T[i] / linalg.norm(Q_t_T[i])
    R[i][i] = linalg.norm(Q_t_T[i])
    for j in range(i+1, n):
      tmp = Q_t_T[j] - Q_T[i] * (Q_t_T[j]@(np.transpose(Q_T[i])))
      Q_t_T[j] = np.copy(tmp)
      tmp = np.zeros(n)
    for k in range(1,i+1):
      R[k-1][i] = AT[i]@np.transpose(Q_T[k-1])
  Q = np.transpose(Q_T)
  return Q,R
```

```python
def hilbert(n):
  A = np.zeros((n, n))
  for i in range(0, n):
    for j in range(0, n):
      A[i][j] = (1 / (i + j + 1))
  return A
```

```python
def Que(A):
  n,n = A.shape
  Q,R = MGS(A)
  AQR_n = linalg.norm(A - Q@R)
  QTQI_n = linalg.norm((np.transpose(Q)@Q) - np.identity(n))

  Q_i, R_i = np.linalg.qr(A)

  print(f'Q from MGS-\n{Q}')
  print(f'\nQ from inbuilt function-\n{Q_i}')
  print(f'\nR from CGS-\n{R}')
  print(f'\nR from inbuilt function-\n{R_i}\n')
  print(f'\n|| A - QR ||f-\n{AQR_n}\n')
  print(f'\n|| QTQ - I ||f-\n{QTQI_n}\n')


n = 3

print("For part a\n")
A = np.random.rand(n, n)
Que(A)

print("For part b\n")
B = np.random.rand(n, n-2)
A = (B@np.transpose(B))
Que(A)

print("For part c\n")
A = A + 0.0001*np.identity(n)
Que(A)

print("For part d\n")
A = 0.0001*np.identity(n) + hilbert(n)
Que(A)
```
