```python
import numpy as np
from numpy import linalg

def CGS(A):
  n,n = A.shape
  Q_T = np.zeros((n,n))
  Q_T[0] = (np.transpose(A))[0]/linalg.norm((np.transpose(A))[0])
  R = np.zeros((n,n))
  R[0][0] = linalg.norm((np.transpose(A))[0])
  sum = np.zeros(n)
  for i in range(1,n):
    for j in range(0,i):
      a = np.transpose(A)[i]@((np.transpose(Q_T))[j])
      sum += Q_T[j]*a
      v_t = (np.transpose(A))[i]-sum
      v_tn = linalg.norm(v_t)
    if v_tn >= 1e-20:
      Q_T[i] = v_t*(1/v_tn)
      R[i][i] = v_tn
    else:
      Q_T[i] = np.zeros(n)
      R[i][i] = 0
  sum = np.zeros(n)
  Q = np.transpose(Q_T)
  return Q,R 
```

```python
def hilbert(n):
  A = np.zeros((n, n))
  for i in range(0, n):
    for j in range(0, n):
      A[i][j] = (1 / (i + j + 1))
  return A
```

```python
def Que(A):
  n,n = A.shape
  Q,R = CGS(A)
  AQR_n = linalg.norm(A - Q@R)
  QTQI_n = linalg.norm((np.transpose(Q)@Q) - np.identity(n))

  Q_i, R_i = np.linalg.qr(A)

  print(f'Q from CGS-\n{Q}')
  print(f'\nQ from inbuilt function-\n{Q_i}')
  print(f'\nR from CGS-\n{R}')
  print(f'\nR from inbuilt function-\n{R_i}\n')
  print(f'\n|| A - QR ||f-\n{AQR_n}\n')
  print(f'\n|| QTQ - I ||f-\n{QTQI_n}\n')


n = 3

print("For part a\n")
A = np.random.rand(n, n)
Que(A)

print("For part b\n")
B = np.random.rand(n, n-2)
A = (B@np.transpose(B))
Que(A)

print("For part c\n")
A = A + 0.0001*np.identity(n)
Que(A)

print("For part d\n")
A = 0.0001*np.identity(n) + hilbert(n)
Que(A)
```
